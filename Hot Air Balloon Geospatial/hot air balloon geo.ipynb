```python
import pandas as pd
```

```python
data = pd.read_csv('20230803205247-79213-data.txt', sep="\t")
data = data[['latitude', 'longitude', 'altitude (m)']]

data.head()
```

```python
# Common imports

import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import numpy as np
import pandas as pd
import os
%matplotlib inline
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create new df(s) for each column from "data" (for easier referencing)
x = data['latitude']
y = data['longitude']
z = data['altitude (m)']

# Set up font family and color for axis titles
font1 = {'family':'serif','color':'darkred'}

fig = plt.figure(figsize = [15, 15])

# Subplot 1
ax1 = fig.add_subplot(1, 3, 1, projection = '3d')
ax1.view_init(0, 90)
ax1.plot3D(x, y, z, 'black')
ax1.locator_params(axis = 'both', nbins = 7)
ax1.set_xlabel('latitude', fontdict = font1)
ax1.set_ylabel('longitude', fontdict = font1)
ax1.set_zlabel('altitude (m)', fontdict = font1);

# Subplot 2
ax3 = fig.add_subplot(1, 3, 2, projection = '3d')
ax3.view_init(35, 45)
ax3.plot3D(x, y, z, 'black')
ax3.locator_params(axis = 'both', nbins = 7)
ax3.set_xlabel('latitude', fontdict = font1)
ax3.set_ylabel('longitude', fontdict = font1)
ax3.set_zlabel('altitude (m)', fontdict = font1);

# Subplot 3
ax4 = fig.add_subplot(1, 3, 3, projection = '3d')
ax4.view_init(25, 75)
ax4.plot3D(x, y, z, 'black')
ax4.locator_params(axis = 'both', nbins = 7)
ax4.set_xlabel('latitude', fontdict = font1)
ax4.set_ylabel('longitude', fontdict = font1)
ax4.set_zlabel('altitude (m)', fontdict = font1);

# not accurate as coordinates and altitude use different metrics, but general shape can be observed
```
